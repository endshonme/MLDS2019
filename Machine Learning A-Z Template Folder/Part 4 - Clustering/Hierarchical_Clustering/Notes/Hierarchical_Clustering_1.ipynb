```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('C:/Users/C5232886/Desktop/MLDS2019/MLDS2019/Machine Learning A-Z Template Folder/Part 4 - Clustering/Hierarchical_Clustering/Mall_Customers.csv')
dataset
```

```python
X = dataset.iloc[:,[3,4]]
X
```

```python
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(X, method='ward'))
# taking each customer as a cluster
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean Distances')
plt.show()
```
